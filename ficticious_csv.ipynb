```python
import csv
import random
from faker import Faker
from datetime import datetime, timedelta

# Initialize Faker library
fake = Faker()

# Convert 'Date of Birth' column to datetime
df['Date_of_Birth'] = pd.to_datetime(df['Date_of_Birth'])

# Calculate age based on 'Date of Birth'
def calculate_age(birth_date):
    current_date = datetime.now()
    age = current_date.year - birth_date.year - ((current_date.month, current_date.day) < (birth_date.month, birth_date.day))
    return age

# Function to generate random date of birth
def generate_dob() -> str:
    start_date = datetime(1950, 1, 1)
    end_date = datetime(2006, 1, 1)
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    return random_date.strftime('%Y-%m-%d')

# Function to generate random monthly income
def generate_income() -> float:
    return round(random.uniform(500.00, 15000.00), 2)

# Generate data and write to CSV
with open('fictitious.csv', 'w', newline='') as csvfile:
    banana = ['Full_Name', 'Date_of_Birth', 'Monthly_Income']
    writer = csv.DictWriter(csvfile, fieldnames=banana)
    writer.writeheader()

    for _ in range(100000):
        full_name = fake.name()
        dob = generate_dob()
        income = generate_income()

        writer.writerow({'Full_Name': full_name, 'Date_of_Birth': dob, 'Monthly_Income': income})


df['Age'] = df['Date_of_Birth'].apply(lambda y: calculate_age(y))

print("CSV file generated successfully.")
```
